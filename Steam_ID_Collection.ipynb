```python
from steam_review_scraper import get_game_ids
import os
import pandas as pd

def save_data_to_excel(data, file_path):
    """Save data to an Excel file."""
    df = pd.DataFrame(data)
    df.to_excel(file_path, index=False)
    print(f'Data saved to {file_path}')

# Example usage
game_ids_data = get_game_ids(1020, filter='relevance')

# Specify the directory where you want to save the Excel file
save_directory = 'path/to/data'

# Create the directory if it does not exist
os.makedirs(save_directory, exist_ok=True)

# Specify the file path where you want to save the Excel file
excel_file_path = os.path.join(save_directory, 'game_ids.xlsx')

# Save the data to an Excel file
save_data_to_excel(game_ids_data, excel_file_path)
```
